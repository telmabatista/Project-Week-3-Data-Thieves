```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
import json
from flatten_json import flatten
```

```python
with open('ticket.json') as f:
   d = json.load(f)

multi = []; inner = {}

def recursive_extract(i):
    global multi, inner

    if type(i) is list:
        if len(i) == 1:
            for k,v in i[0].items():
                if type(v) in [list, dict]:
                    recursive_extract(v)
                else:                
                    inner[k] = v
        else:
            multi = i

    if type(i) is dict:
        for k,v in i.items():
            if type(v) in [list, dict]:
                recursive_extract(v)
            else:                
                inner[k] = v

recursive_extract(d["_embedded"])

data_dict = []
for i in multi:    
    tmp = inner.copy()
    tmp.update(i)
    data_dict.append(tmp)


dict_flat = [flatten(d) for d in data_dict]

df = pd.DataFrame(dict_flat)
df.head()
```
