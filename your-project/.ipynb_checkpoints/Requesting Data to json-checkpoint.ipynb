```python
multi = []; inner = {}
def recursive_extract(i):
    global multi, inner 

    if type(i) is list:
        if len(i) == 1:
            for k,v in i[0].items():
                if type(v) in [list, dict]:
                    recursive_extract(v)
                else:                
                    inner[k] = v
        else:
            multi = i

    if type(i) is dict:
        for k,v in i.items():
            if type(v) in [list, dict]:
                recursive_extract(v)
            else:                
                inner[k] = v
                

def country_df(country):
    inputs = {"apikey" : key,
          "size" : 200,
          "countryCode" : country}

    r = requests.get(url,inputs)
    d = r.json()
    
    recursive_extract(d["_embedded"])

    data_dict = []
    for i in multi:    
        tmp = inner.copy()
        tmp.update(i)
        data_dict.append(tmp)


    dict_flat = [flatten(d) for d in data_dict]

    df = pd.DataFrame(dict_flat)

    df = df[columns_keep]

    df.columns = name_columns
    
    if df.iloc[0]["currency"] == "EUR":
        df["min_price_eur"] = df["min_price"]
        df["max_price_eur"] = df["max_price"]
    
    else:
        cur = "rates." + df.iloc[0]["currency"]
        df["min_price_eur"] = df["min_price"] / exchange.loc[cur][0]
        df["max_price_eur"] = df["max_price"] / exchange.loc[cur][0]
    
    return df
```

```python
frames = []

for c in countries:
    try:
        frames.append(country_df(c))
    except:
        pass
    
result = pd.concat(frames)

result.head()

result.to_csv("ticket_df.csv",index=False)
```
